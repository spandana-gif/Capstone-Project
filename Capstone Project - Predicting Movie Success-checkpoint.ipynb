### Outlier Treatment
 -- I am going to check for outliers using the IQR method

### Interquartile Range (IQR)
IQR = Q3 ‚àí Q1

where:

- Q1 (25th percentile) is the first quartile
- Q3 (75th percentile) is the third quartile

### Detecting Outliers
IQR is often used for outlier detection:

- Lower bound = ùëÑ1 ‚àí 1.5 √ó ùêºùëÑùëÖ
- Upper bound = Q3 + 1.5 √ó IQR


```python
# Define the numerical features you want to check
num_cols = ['budget', 'duration', 'num_voted_users', 'gross', 'imdb_score', 'num_critic_for_reviews', 'num_user_for_reviews', 'cast_total_facebook_likes', 'director_facebook_likes','actor_1_facebook_likes', 'movie_facebook_likes']  
# Loop through each column to detect outliers
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\n{col} - Number of outliers: {outliers.shape[0]}")
```

```python
num_cols = ['budget', 'duration', 'num_voted_users', 'gross', 'imdb_score',
            'num_critic_for_reviews', 'num_user_for_reviews',
            'cast_total_facebook_likes', 'director_facebook_likes', 'actor_1_facebook_likes', 'movie_facebook_likes']

# Replace NaNs with median
for col in num_cols:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)
```

### Predicting Movie Success with Cross-Validation & Hyperparameter Tuning

```python
from sklearn.model_selection import cross_val_score 

# Instantiate model
rfc = RandomForestClassifier(random_state=42)

# Perform 5-fold cross-validation on training data
cv_scores = cross_val_score(rfc, X_train, y_train, cv=5, scoring='accuracy')

print("Cross-validation scores:", cv_scores)
print("Mean CV Accuracy:", cv_scores.mean())
```

```python
# Cross-Validation Visualization

plt.figure(figsize=(8, 5))
plt.plot(range(1, 6), cv_scores, marker='o', linestyle='--', color='green')
plt.title('Cross-Validation Accuracy per Fold')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.ylim(0.7, 1.0)
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

# Initialize model
rfc = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=rfc,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

# Fit to training data
grid_search.fit(X_train, y_train)

# Best model and score
print("Best Parameters:", grid_search.best_params_)
print("Best CV Accuracy:", grid_search.best_score_)
```

```python
# Hyperparameter Tuning Visualization
results_df = pd.DataFrame(grid_search.cv_results_)
plt.figure(figsize=(10, 5))
plt.plot(results_df['param_n_estimators'], results_df['mean_test_score'], marker='o', linestyle='-', color='green')
plt.title('Accuracy vs n_estimators')
plt.xlabel('n_estimators')
plt.ylabel('Mean CV Accuracy')
plt.grid(True)
plt.show()
```

## Evaluate on Test Set

```python
# Final model evaluation on test set
final_model = grid_search.best_estimator_
y_final_pred = final_model.predict(X_test)
print("\nFinal Model Evaluation:\n")
print(classification_report(y_test, y_final_pred, target_names=['Flop', 'Avg', 'Hit']))
```
